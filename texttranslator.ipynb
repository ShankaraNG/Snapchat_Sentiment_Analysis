```python
import pandas as pd
import regex as re
from deep_translator import GoogleTranslator
import time
```

```python
data_df = pd.read_csv(r'E:\PythonCertificates\Snapchat Sentiment Analysis\archive\snapchat_reviews.csv')
data_df.head(10)
```

```python
def clean_text_embeddings(text):
    text = str(text).lower()
    text = re.sub(r'[^\x00-\x7F]', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)
    text = re.sub(r'\{.*?\}|\[.*?\]|<.*?>|\(.*?\)', '', text)
    text = re.sub(r'[\U00010000-\U0010ffff]', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
data_df['cleaned_content'] = data_df['content'].apply(clean_text_embeddings)
```

```python
data_df = data_df.drop_duplicates(subset=['cleaned_content'])
```

```python
def translate_to_english(text):
    try:
        # lang = detect_language(text)
        # if lang == 'en':
        #     return text
        translated = GoogleTranslator(source='auto', target='en').translate(text)
        time.sleep(0.2)
        return translated
    except:
        return text
```

```python
data_df['cleaned_content_translated'] = data_df['cleaned_content'].apply(translate_to_english)
```

```python
data_df.to_csv(r'E:\PythonCertificates\Snapchat Sentiment Analysis\archive\snapchat_reviews_translated.csv', index=False, encoding='utf-8-sig')
```
